```python
import networkx as nx
import numpy as np
```

```python
print(nx.__version__)
```

```python
n = 4
edges = [(i, j) for i in range(n) for j in range(0, i)]

g = nx.Graph(edges)
nx.draw(g)
```

```python
n = 5
edges = [(i, j) for i in range(n) for j in range(0, i)]

g = nx.Graph(edges)
nx.draw(g)
```

```python
n = 5
edges = [(i, j) for i in range(0, 2) for j in range(2, 5)]
g = nx.Graph(edges)
pos = nx.drawing.layout.bipartite_layout(g, [0, 1])
nx.draw(g, pos = pos)
```

```python
n = 6
edges = [(i, j) for i in range(0, 3) for j in range(3, 6)]
g = nx.Graph(edges)
pos = nx.drawing.layout.bipartite_layout(g, [0, 1, 2])
nx.draw(g, pos = pos)
```

```python
g = nx.hypercube_graph(4)
nx.draw(g)
```

```python
g = nx.ladder_graph(5)
nx.draw(g)
```

```python
g = nx.barbell_graph(20, 1)
nx.draw(g)
```

```python
g = nx.wheel_graph(100)
nx.draw(g)
```
